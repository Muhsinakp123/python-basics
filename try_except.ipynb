```python
# Division Calculator
# Write a program that asks the user for two numbers and prints their division result

num1_str = (input("Enter the First Number"))
num2_str = (input("Enter the Second Number"))
try:
    
    num1 = int(num1_str)
    num2 = int(num2_str)
    
    # Perform division
    result = num1 / num2
    print(f"Result of dividing {num1} by {num2} is: {result}")

except ValueError:
    print("Invalid input! Please enter numeric values only.")
except ZeroDivisionError:
    print("Division by zero is not allowed.")
```

```python
# List Index Finder
# Ask the user to enter an index number

numbers = [10, 20, 30, 40]
index_str = input("Enter the Index Value")
try:
    index = int(index_str)
    value = numbers[index]
    print(f"The Index of {index} is {value} ")
except ValueError:
    print("Invalid input! Please enter numeric values only.")
except IndexError:
    print("The index is out of Range")
    
```

```python
# File Reader
# Ask the user for a filename and try to open it

file_name = input("Enter Your File Name")
try:
    with open (file_name,"r") as file:
        content = file.read()
        print("----File contents -----")
        print(content)
except FileNotFoundError:
    print(f"The file '{file_name}' does not exist.")
```
